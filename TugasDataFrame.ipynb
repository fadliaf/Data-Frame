```python
import pandas as pd
pd.set_option ('display.max_rows', None)
dataframe = pd.read_csv('data_sampah.csv')

dataframe
```

```python
total_produksi = {}

jumlah = 0
for index, row in dataframe.iterrows():
    tahun = row['tahun']
    produksi = row['jumlah_produksi_sampah']
    if tahun == 2020:   
        jumlah += produksi   

total_produksi[2020] = jumlah

print(f"Total produksi sampah di Jawa Barat pada tahun 2020 adalah {total_produksi[2020]}")
```

```python
total_produksi= {}

for index, row in dataframe.iterrows():
    tahun = row['tahun']  
    produksi = row['jumlah_produksi_sampah']  
    
    if tahun in total_produksi:
        total_produksi[tahun] += produksi
    else:
        total_produksi[tahun] = produksi

for tahun, total in total_produksi.items():
    print(f"Pada tahun {tahun} total produksi sampahnya adalah {total:.2f}")
```

```python
total_produksi = {}

for index, row in dataframe.iterrows():
    kota = row['nama_kabupaten_kota'] 
    produksi = row['jumlah_produksi_sampah']  
    
    if kota in total_produksi:
        total_produksi[kota] += produksi
    else:
        total_produksi[kota] = produksi

total_produksi
# for kota, total in total_produksi.items():
#     print(f"Di {kota} memproduksi sampah sebanyak {total:.2f}")
```

```python
df_sampah = pd.DataFrame(total_produksi)
df_sampah.to_csv('produski_sampah.csv', index=False)
df_sampah.to_excel('produski_sampah.xlsx', index=False)
```

```python
df_sampah = pd.DataFrame(list(total_produksi.items()), columns=['Kota/Kabupaten', 'Total (ton)'])

import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))
plt.barh(df_sampah['Kota/Kabupaten'], df_sampah['Total (ton)'], color='skyblue')
plt.xlabel('Jumlah Produksi Sampah (ton)')
plt.title('Jumlah Produksi Sampah (Kota/Kabupaten)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
def kategori(total):
    if total > 400:
        return 'sampah harus segera ditanggulangi'
    else:
        return 'sampah masih dalam kategori aman'
    
df_sampah['Kategori'] = df_sampah['Total (ton)'].apply(kategori)
df_sampah
```

```python
harga = 123000

df_sampah['Bayaran'] = df_sampah['Total (ton)'] * harga
df_sampah
```

```python
def pajak(row):
    kategori = row['Kategori']
    harga = row['Bayaran']
    if kategori == 'sampah harus segera ditanggulangi':
        return harga * 0.05

df_sampah['Pajak'] = df_sampah.apply(pajak, axis=1)
df_sampah
```
